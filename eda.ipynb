Exploring various fingerprint datasets...

```python
import glob
import matplotlib.pyplot
import os.path
import random

datasets_path = os.path.abspath("./datasets/")
```

```python
def show_random_images(image_paths):
    shown_image_paths = random.sample(image_paths, 20)

    figure, axes = matplotlib.pyplot.subplots(nrows=2, ncols=4, figsize=(12, 6))

    for i, axis in enumerate(axes.flat):
        image = matplotlib.pyplot.imread(shown_image_paths[i])
        axis.imshow(image)
        axis.set_title(os.path.basename(shown_image_paths[i]))

    matplotlib.pyplot.tight_layout(True)
    matplotlib.pyplot.show()
```

## SOCOFing Dataset

```python
real_images_glob = os.path.join(datasets_path, "SOCOFing/Real/*.BMP")
real_image_paths = glob.glob(real_images_glob)

len(real_image_paths)
```

```python
show_random_images(real_image_paths)
```

```python
altered_images_glob = os.path.join(datasets_path, "SOCOFing/Altered/Altered-Medium/*.BMP")
altered_image_paths = glob.glob(altered_images_glob)

len(altered_image_paths)
```

```python
# todo: figure out the weird colors
show_random_images(altered_image_paths)
```
